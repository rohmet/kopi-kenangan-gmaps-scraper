### Cell 1: Import Libraries

```python
import pandas as pd
import os
import glob
```

### Cell 2: Setup Path & File Discovery

```python
path = './data-scrap/'
all_files = glob.glob(os.path.join(path, "*.csv"))
print(f"Jumlah file CSV ditemukan: {len(all_files)}")
```

### Cell 3: Data Merging

```python
df_list = []
for file in all_files:
    try:
        df = pd.read_csv(file)
        # Ambil nama gerai dari nama file tanpa ekstensi .csv
        nama_gerai = os.path.splitext(os.path.basename(file))[0]
        # Tambahkan kolom nama_gerai
        df['nama_gerai'] = nama_gerai
        df_list.append(df)
    except Exception as e:
        print(f"Error reading {file}: {e}")

df_combined = pd.concat(df_list, ignore_index=True)
print(f"Ukuran DataFrame gabungan: {df_combined.shape}")
```

### Cell 4: Initial Data Inspection

```python
display(df_combined.head())
print("\n--- Info DataFrame ---")
df_combined.info()
print("\n--- Missing Values ---")
print(df_combined.isnull().sum())
```

### Cell 5: Column Renaming & Normalization

```python
df_combined.columns = df_combined.columns.str.lower().str.replace(' ', '_')
print("Kolom setelah dinormalisasi:")
print(df_combined.columns.tolist())
```

### Cell 6: Data Deduplication

```python
initial_rows = len(df_combined)
df_combined = df_combined.drop_duplicates(subset=['nama_pengulas', 'ulasan'])
final_rows = len(df_combined)
print(f"Baris sebelum: {initial_rows}")
print(f"Baris sesudah (tanpa duplikat): {final_rows}")
print(f"Jumlah baris duplikat dihapus: {initial_rows - final_rows}")
```

### Cell 7: Handling Missing Values

```python
df_combined = df_combined.dropna(subset=['ulasan', 'rating'])
print(f"Ukuran setelah menghapus NaN: {df_combined.shape}")
```

### Cell 8: Final Data Validation

```python
print("--- Distribusi Rating ---")
print(df_combined['rating'].value_counts())
print("\n--- Final Info ---")
df_combined.info()
```

### Cell 9: Data Export

```python
output_dir = './data/processed/'
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, 'cleaned_data.csv')
df_combined.to_csv(output_path, index=False)
print(f"Data akhir berhasil disimpan ke: {output_path}")
```
